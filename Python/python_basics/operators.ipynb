```python
# simple calculator

value1 = int(input("Enter first number: "))
value2 = int(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ")

if operation == '+':
    result = value1 + value2
elif operation == '-':
    result = value1 - value2
elif operation == '*':
    result = value1 * value2
elif operation == '/':
    if value2 != 0:
        result = value1 / value2
    else:
        result = "Error: Division by zero"
else:
    result = "Invalid operation"        
print("Result:", result)
```
