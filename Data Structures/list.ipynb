```python
# list is a data structure in Python that is mutable, ordered, and allows duplicate elements. It is defined using square brackets [] and can contain elements of different data types. Lists are commonly used to store collections of items and provide various methods for manipulation, such as adding, removing, and sorting elements.

list = [1, 2, 3, "Hello", 4.5, True]

print(list)
```

```python
# lists are mutable and ordered, which means you can change their contents and they maintain the order of elements. You can perform various operations on lists, such as:

# 1. Adding elements:
list.append("New Element")  # Adds an element to the end of the list
print(list) 

list.insert(2, "Inserted Element")  # Inserts an element at a specific index
print(list)

# 2. Removing elements:
list.remove("Inserted Element")  # Removes the first occurrence of the specified element
print(list)


print(list.pop(1))  # Removes and returns the element at the specified index
print(list)


catch = list[0]
print(catch)  # Accessing an element by index
```

```python
list = ["grapes" , "oranges", "apples", "grapes", "bananas"]

print(list[-1])
print(list[len(list) - 1])  # Accessing the last element of the list

# list slicing allows you to access a range of elements in a list. The syntax for slicing is list[start:stop:step], where start is the index to begin slicing, stop is the index to end slicing (exclusive), and step is the interval between elements. 


print(list[:])

print(list[::2])  # Accessing every second element in the list


# lists are mutable

list[0] = "kiwi"  # Modifying an element in the list
print(list)

print(list.count("grapes"))  # Counting occurrences of an element in the list


print(list.index("oranges"))  # Finding the index of the first occurrence of an element in the list

print(list.sort())  # Sorting the list in place
print(list)

print(sorted(list))  # Sorting the list and returning a new sorted list

print(list)  # Original list remains unchanged after using sorted()

print(list.reverse())  # Reversing the list in place
print(list)

print(list[::-1])  # Reversing the list using slicing   

print(list)  # Original list remains unchanged after using slicing to reverse

print(list[::-3])
```

```python
# Methods like append(), insert(), remove(), pop(), count(), index(), sort(), and reverse() are commonly used to manipulate lists in Python. Additionally, slicing allows for flexible access to elements in a list.


list = [ 4 , 5 , 6 , 7 , 8 , 9 , 10 ]

list[1:5] = [50, 60, 70]  # Replacing a slice of the list with new values

print(list)


list.clear()

list.append("New Element")

print(list)
```

```python
# list comprehensions provide a concise way to create lists. They consist of brackets containing an expression followed by a for clause, and optionally, one or more if clauses. The expression is evaluated for each item in the iterable, and the resulting values are collected into a new list.
number_list = [i for i in range(10)]  # Creating a list of numbers from 0 to 9 using a list comprehension

print(number_list)

number_list = [i**2 for i in range(10)]  # Creating a list of squares of numbers from 0 to 9 using a list comprehension

print(number_list)


number_list = [i for i in range(10) if i % 2 == 0]  # Creating a list of even numbers from 0 to 9 using a list comprehension with a condition

print(number_list)

number_list = [i**2 for i in range(10) if i % 2 == 0]  # Creating a list of squares of even numbers from 0 to 9 using a list comprehension with a condition

print(number_list)

number_list = [i**2 for i in range(10) if i % 2 == 0 and i > 5]  # Creating a list of squares of even numbers greater than 5 from 0 to 9 using a list comprehension with multiple conditions

print(number_list)
```
